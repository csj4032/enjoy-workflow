```python
import json
import logging
import random
import time
import uuid
from datetime import datetime, UTC
from typing import Optional, Dict

from confluent_kafka import Producer
from confluent_kafka.admin import AdminClient, NewTopic
from confluent_kafka.error import KafkaException
from faker import Faker
```

```python
def get_admin(kafka_config: dict) -> AdminClient:
    return AdminClient(kafka_config)


def list_topics(admin: AdminClient, timeout: float = 10.0):
    md = admin.list_topics(timeout=timeout)
    topics = md.topics.keys()
    for t in sorted(topics):
        print(t)


def create_topic(admin: AdminClient, topic_name: str, partitions: int = 1, replication_factor: int = 1, timeout: float = 30.0):
    new_topic = NewTopic(topic_name, num_partitions=partitions, replication_factor=replication_factor)
    fs = admin.create_topics([new_topic], request_timeout=timeout)
    try:
        fs[topic_name].result()
        logging.info(f"created topic: {topic_name}")
    except KafkaException as e:
        msg = str(e)
        if "TOPIC_ALREADY_EXISTS" in msg or "TopicExistsError" in msg:
            logging.info(f"topic already exists: {topic_name}")
        else:
            raise RuntimeError(f"create_topics failed for {topic_name}: {e}") from e


def delete_topics(admin: AdminClient, topics_to_delete, timeout: float = 30.0) -> None:
    if isinstance(topics_to_delete, str):
        topics_to_delete = [topics_to_delete]
    topics_to_delete = list(topics_to_delete)
    logging.info("Deleting topics:", topics_to_delete)
    fs = admin.delete_topics(topics_to_delete, request_timeout=timeout)
    for t in topics_to_delete:
        try:
            fs[t].result()
            logging.info(f"deleted topic: {t}")
        except KafkaException as e:
            raise RuntimeError(f"delete_topics failed for {t}: {e}") from e


def _delivery_report(err, msg) -> None:
    if err is not None:
        logging.info(f"Delivery failed: {err}")
    else:
        logging.info(f"Delivered to {msg.topic()} [{msg.partition()}] @ offset {msg.offset()}")


def get_producer(kafka_config):
    return Producer({**kafka_config, "acks": "all", "retries": 1, "request.timeout.ms": 1500, "connections.max.idle.ms": 2000})


def send_json(producer_: Producer, topic: str, key: Optional[str], value: Dict, flush: bool = False) -> None:
    key_bytes = key.encode("utf-8") if key else None
    value_bytes = json.dumps(value, ensure_ascii=False).encode("utf-8")
    producer_.produce(topic=topic, key=key_bytes, value=value_bytes, on_delivery=_delivery_report)
    producer_.poll(0)
    if flush:
        producer_.flush(10)


def generate_product(fake_: Faker) -> dict:
    return {
        "product_id": str(uuid.uuid4()),
        "name": fake_.catch_phrase(),
        "category": random.choice(["전자기기", "생활용품", "식품", "도서", "의류", "스포츠", "뷰티"]),
        "price": random.randint(1_000, 500_000),
        "currency": "KRW",
        "stock": random.randint(0, 500),
        "created_at": datetime.now(UTC).isoformat()
    }
```

```python
_kafka_config = {
    "bootstrap_servers": "ec2-13-209-6-40.ap-northeast-2.compute.amazonaws.com:29092,ec2-13-209-6-40.ap-northeast-2.compute.amazonaws.com:29093,ec2-13-209-6-40.ap-northeast-2.compute.amazonaws.com:29094",
    "security_protocol": "PLAINTEXT"
}
_topic_name = "mmix-products-topic"
```

```python
create_topic(get_admin(_kafka_config), "mmix-product-topic")
list_topics(get_admin(_kafka_config))
```
