```python
import pymysql

from airflow.settings import Session
from airflow.models.connection import Connection

from airflow.models import Variable
from airflow.providers.amazon.aws.hooks.s3 import S3Hook
from airflow.providers.mysql.hooks.mysql import MySqlHook
```

```python
conn_id = Variable.get("mmix-mysql-primary-mmix-conn-id")
s = Session()
c = s.query(Connection).filter(Connection.conn_id == conn_id).one()
print("FOUND:", c.conn_type, c.host, c.schema, c.login, c.port)
```

```python
db = pymysql.connect(host=c.host, user=c.login, password=c.password, database=c.schema,port=c.port or 3306, connect_timeout=5)

with db.cursor() as cur:
    cur.execute("SELECT 1")
    print("MYSQL OK:", cur.fetchone())
db.close()
```
